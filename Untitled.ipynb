```python
from pathlib import Path
import random

from PIL import Image

import torch
from torch.utils.data import Dataset, DataLoader
from torchvision.transforms import functional as TF
```

```python
DATASET_NAME = "Fantasy"

PROCESSED_ROOT = Path("processedDatasets") / DATASET_NAME

IMAGE_EXTENSIONS = {".webp",}

BATCH_SIZE = 32
NUM_WORKERS = 2
```

```python
def scale_0_1_to_neg1_1(x: torch.Tensor) -> torch.Tensor:
    """
    Convert image tensor from [0, 1] to [-1, 1].

    Input shape:  (C, H, W)
    Output shape: (C, H, W)
    """
    return x * 2.0 - 1.0
```

```python
class ProcessedImageDataset(Dataset):
    """
    Dataset for flattened processed image folders.

    __getitem__ returns:

        original:
            torch.float32 tensor in [0, 1]
            shape: (3, H, W)

        lpips_ready:
            torch.float32 tensor in [-1, 1]
            shape: (3, H, W)
            has 50% chance of horizontal flip
    """

    def __init__(
        self,
        root: str | Path,
        image_extensions: set[str] = IMAGE_EXTENSIONS,
        horizontal_flip_prob: float = 0.5,
    ):
        self.root = Path(root)
        self.image_extensions = {ext.lower() for ext in image_extensions}
        self.horizontal_flip_prob = horizontal_flip_prob

        if not self.root.exists():
            raise FileNotFoundError(f"Dataset folder does not exist: {self.root}")

        self.image_paths = sorted(
            p for p in self.root.iterdir()
            if p.is_file() and p.suffix.lower() in self.image_extensions
        )

        if len(self.image_paths) == 0:
            raise ValueError(f"No images found in: {self.root}")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, index: int):
        path = self.image_paths[index]

        with Image.open(path) as img:
            img = img.convert("RGB")

            # Converts PIL image to float32 torch tensor in [0, 1],
            # with shape (C, H, W).
            original = TF.to_tensor(img).float()

        augmented = original.clone()

        if random.random() < self.horizontal_flip_prob:
            augmented = TF.hflip(augmented)

        lpips_ready = scale_0_1_to_neg1_1(augmented)

        return original, lpips_ready
```

```python
@dataclass
class VQGANConfig:
    image_height: int = 256
    image_width: int = 192
    image_channels: int = 3

    # Your requested compression factor.
    f: int = 4

    # Your requested codebook embedding dimension.
    z_channels: int = 3

    # Number of discrete codebook entries.
    # This is separate from codebook dimension.
    n_embed: int = 256

    # VQ-VAE/VQGAN commitment weight, beta in the paper-style codebook loss.
    beta: float = 0.25

    # Lightweight architecture.
    base_channels: int = 96
    channel_multipliers: tuple = (1, 2)

    # Loss weights.
    perceptual_weight: float = 1.0
    reconstruction_weight: float = 1.0
    codebook_weight: float = 1.0

    # Adversarial settings.
    disc_start_step: int = 5_000
    disc_weight: float = 0.8
    disc_factor: float = 1.0
    adaptive_weight_max: float = 1e4

    # Optimizer.
    lr: float = 4.5e-6
    betas: tuple = (0.5, 0.9)

    # Training.
    epochs: int = 100
    batch_size: int = 32
    num_workers: int = 4
    grad_accum_steps: int = 1

    # Logging/checkpointing.
    sample_every_steps: int = 500
    ckpt_every_steps: int = 2_000
    out_dir: str = "runs/vqgan_fantasy"


cfg = VQGANConfig()

assert cfg.image_height % cfg.f == 0
assert cfg.image_width % cfg.f == 0

latent_h = cfg.image_height // cfg.f
latent_w = cfg.image_width // cfg.f

print("Latent grid:", latent_h, "x", latent_w)
print("Latent channels/codebook dimension:", cfg.z_channels)
print("Tokens per image:", latent_h * latent_w)
```

```python
def to_neg1_1(x_0_1: torch.Tensor) -> torch.Tensor:
    return x_0_1 * 2.0 - 1.0


def to_0_1(x_neg1_1: torch.Tensor) -> torch.Tensor:
    return ((x_neg1_1 + 1.0) / 2.0).clamp(0.0, 1.0)


def adopt_weight(weight: float, global_step: int, threshold: int, value: float = 0.0) -> float:
    """
    VQGAN-style delayed discriminator activation.
    Before disc_start_step, adversarial weight is zero.
    """
    if global_step < threshold:
        return value
    return weight


def hinge_d_loss(logits_real: torch.Tensor, logits_fake: torch.Tensor) -> torch.Tensor:
    """
    Hinge discriminator loss:
        mean(relu(1 - D(real))) + mean(relu(1 + D(fake)))
    """
    loss_real = torch.mean(F.relu(1.0 - logits_real))
    loss_fake = torch.mean(F.relu(1.0 + logits_fake))
    return 0.5 * (loss_real + loss_fake)


def vanilla_g_loss(logits_fake: torch.Tensor) -> torch.Tensor:
    """
    Generator adversarial loss used in the common VQGAN implementation:
        -mean(D(fake))
    """
    return -torch.mean(logits_fake)


def calculate_adaptive_weight(
    nll_loss: torch.Tensor,
    g_loss: torch.Tensor,
    last_layer: torch.Tensor,
    max_weight: float = 1e4,
    eps: float = 1e-4,
) -> torch.Tensor:
    """
    Adaptive discriminator weight from the VQGAN training recipe.

    Conceptually:
        lambda = ||grad_L_rec|| / (||grad_L_GAN|| + eps)

    The original VQGAN implementation computes this ratio with respect to the
    decoder's last layer.
    """
    nll_grads = torch.autograd.grad(
        nll_loss,
        last_layer,
        retain_graph=True,
        create_graph=True,
        allow_unused=True,
    )[0]

    g_grads = torch.autograd.grad(
        g_loss,
        last_layer,
        retain_graph=True,
        create_graph=True,
        allow_unused=True,
    )[0]

    if nll_grads is None or g_grads is None:
        return torch.tensor(0.0, device=nll_loss.device)

    d_weight = torch.norm(nll_grads) / (torch.norm(g_grads) + eps)
    d_weight = torch.clamp(d_weight, 0.0, max_weight).detach()
    return d_weight
```

```python
class ResBlock(nn.Module):
    def __init__(self, channels: int):
        super().__init__()

        self.block = nn.Sequential(
            nn.GroupNorm(num_groups=min(32, channels), num_channels=channels),
            nn.SiLU(),
            nn.Conv2d(channels, channels, kernel_size=3, padding=1),
            nn.GroupNorm(num_groups=min(32, channels), num_channels=channels),
            nn.SiLU(),
            nn.Conv2d(channels, channels, kernel_size=3, padding=1),
        )

    def forward(self, x):
        return x + self.block(x)


class Downsample(nn.Module):
    def __init__(self, in_channels: int, out_channels: int):
        super().__init__()
        self.conv = nn.Conv2d(
            in_channels,
            out_channels,
            kernel_size=4,
            stride=2,
            padding=1,
        )

    def forward(self, x):
        return self.conv(x)


class Upsample(nn.Module):
    def __init__(self, in_channels: int, out_channels: int):
        super().__init__()

        self.net = nn.Sequential(
            nn.Upsample(scale_factor=2, mode="nearest"),
            nn.Conv2d(
                in_channels,
                out_channels,
                kernel_size=3,
                stride=1,
                padding=1,
            ),
        )

    def forward(self, x):
        return self.net(x)
```

```python
class Encoder(nn.Module):
    def __init__(
        self,
        image_channels: int = 3,
        base_channels: int = 96,
        z_channels: int = 3,
    ):
        super().__init__()

        ch = base_channels

        self.net = nn.Sequential(
            nn.Conv2d(image_channels, ch, kernel_size=3, padding=1),

            ResBlock(ch),
            Downsample(ch, ch * 2),      # 256x192 -> 128x96

            ResBlock(ch * 2),
            Downsample(ch * 2, ch * 4),  # 128x96 -> 64x48

            ResBlock(ch * 4),
            ResBlock(ch * 4),

            nn.GroupNorm(num_groups=min(32, ch * 4), num_channels=ch * 4),
            nn.SiLU(),
            nn.Conv2d(ch * 4, z_channels, kernel_size=1),
            
            # new 1x1 conv layer
        )

    def forward(self, x):
        return self.net(x)
```

```python
class Decoder(nn.Module):
    def __init__(
        self,
        image_channels: int = 3,
        base_channels: int = 96,
        z_channels: int = 3,
    ):
        super().__init__()

        ch = base_channels
        # new 1x1 conv layer

        self.conv_in = nn.Conv2d(z_channels, ch * 4, kernel_size=3, padding=1)

        self.mid = nn.Sequential(
            ResBlock(ch * 4),
            ResBlock(ch * 4),
        )

        self.up = nn.Sequential(
            Upsample(ch * 4, ch * 2),    # 64x48 -> 128x96
            ResBlock(ch * 2),

            Upsample(ch * 2, ch),        # 128x96 -> 256x192
            ResBlock(ch),

            nn.GroupNorm(num_groups=min(32, ch), num_channels=ch),
            nn.SiLU(),
        )

        self.conv_out = nn.Conv2d(ch, image_channels, kernel_size=3, padding=1)

        

    def forward(self, z):
        
        # use 1x1 conv layer
        h = self.conv_in(z)
        h = self.mid(h)
        h = self.up(h)
        x_hat = self.conv_out(h)
        return torch.tanh(x_hat)

    def get_last_layer(self):
        return self.conv_out.weight
```

```python
class VectorQuantizer(nn.Module):
    """
    Vector quantizer.

    z shape:
        (B, C, H, W)

    C must equal embedding_dim.
    """

    def __init__(self, n_embed: int, embedding_dim: int, beta: float = 0.25):
        super().__init__()

        self.n_embed = n_embed
        self.embedding_dim = embedding_dim
        self.beta = beta

        self.embedding = nn.Embedding(n_embed, embedding_dim)
        '''self.embedding.weight.data.uniform_(
            -1.0 / n_embed,
            1.0 / n_embed,
        )'''
        self.embedding.weight.data.uniform_(
            -1.0,
            1.0,
        )

    def forward(self, z):
        B, C, H, W = z.shape
        assert C == self.embedding_dim

        # BCHW -> BHWC -> flat
        z_perm = einx.id("b c h w -> b h w c", z).contiguous()
        z_flat = z_perm.view(-1, C)

        # Squared L2 distance to codebook entries.
        distances = (
            torch.sum(z_flat ** 2, dim=1, keepdim=True)
            + torch.sum(self.embedding.weight ** 2, dim=1)
            - 2.0 * torch.matmul(z_flat, self.embedding.weight.t())
        )

        encoding_indices = torch.argmin(distances, dim=1)
        z_q_flat = self.embedding(encoding_indices)

        z_q = z_q_flat.view(B, H, W, C)

        # Paper-style VQ loss:
        # ||sg[z] - z_q||^2 + beta * ||z - sg[z_q]||^2
        codebook_loss = F.mse_loss(z_q, z_perm.detach())
        commitment_loss = F.mse_loss(z_perm, z_q.detach())
        vq_loss = codebook_loss + self.beta * commitment_loss

        # Straight-through estimator.
        z_q = z_perm + (z_q - z_perm).detach()

        # BHWC -> BCHW
        z_q = einx.id("b h w c -> b c h w", z_q).contiguous()

        encoding_indices = encoding_indices.view(B, H, W)

        return z_q, vq_loss, encoding_indices

    @torch.no_grad()
    def codebook_usage(self, indices: torch.Tensor) -> int:
        return indices.unique().numel()
```

```python
class VQGANAutoencoder(nn.Module):
    def __init__(self, cfg: VQGANConfig):
        super().__init__()

        self.encoder = Encoder(
            image_channels=cfg.image_channels,
            base_channels=cfg.base_channels,
            z_channels=cfg.z_channels,
        )

        # Encoder latent -> codebook latent space.
        self.quant_conv = nn.Conv2d(
            cfg.z_channels,
            cfg.z_channels,
            kernel_size=1,
        )

        self.quantizer = VectorQuantizer(
            n_embed=cfg.n_embed,
            embedding_dim=cfg.z_channels,
            beta=cfg.beta,
        )

        # Codebook latent space -> decoder latent space.
        self.post_quant_conv = nn.Conv2d(
            cfg.z_channels,
            cfg.z_channels,
            kernel_size=1,
        )

        self.decoder = Decoder(
            image_channels=cfg.image_channels,
            base_channels=cfg.base_channels,
            z_channels=cfg.z_channels,
        )

    def encode(self, x, return_prequant: bool = False):
        z = self.encoder(x)

        z = self.quant_conv(z)

        # experimental fix for your collapse.
        z = torch.tanh(z)

        z_q, vq_loss, indices = self.quantizer(z)

        return z_q, vq_loss, indices

    def decode(self, z_q):
        z_q = self.post_quant_conv(z_q)
        return self.decoder(z_q)

    def forward(self, x):
        z_q, vq_loss, indices = self.encode(x)
        x_hat = self.decode(z_q)
        return x_hat, vq_loss, indices

    def get_last_layer(self):
        return self.decoder.get_last_layer()
```

```python
class PatchDiscriminator(nn.Module):
    def __init__(self, image_channels: int = 3, base_channels: int = 64, n_layers: int = 3):
        super().__init__()

        layers = []

        in_ch = image_channels
        out_ch = base_channels

        layers.append(
            nn.Conv2d(in_ch, out_ch, kernel_size=4, stride=2, padding=1)
        )
        layers.append(nn.LeakyReLU(0.2, inplace=True))

        in_ch = out_ch

        for i in range(1, n_layers):
            out_ch = min(base_channels * (2 ** i), 512)

            layers.append(
                nn.Conv2d(
                    in_ch,
                    out_ch,
                    kernel_size=4,
                    stride=2,
                    padding=1,
                    bias=False,
                )
            )
            layers.append(nn.BatchNorm2d(out_ch))
            layers.append(nn.LeakyReLU(0.2, inplace=True))

            in_ch = out_ch

        layers.append(
            nn.Conv2d(
                in_ch,
                1,
                kernel_size=3,
                stride=1,
                padding=1,
            )
        )

        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)
```

```python
autoencoder = VQGANAutoencoder(cfg).to(device)
discriminator = PatchDiscriminator(
    image_channels=cfg.image_channels,
    base_channels=64,
    n_layers=3,
).to(device)

if lpips is None:
    raise ImportError("Install lpips first: pip install lpips")

perceptual_loss_fn = lpips.LPIPS(net="vgg").to(device)
perceptual_loss_fn.eval()

# LPIPS is used as a frozen loss network.
for p in perceptual_loss_fn.parameters():
    p.requires_grad = False

opt_ae = torch.optim.Adam(
    autoencoder.parameters(),
    lr=cfg.lr,
    betas=cfg.betas,
)

opt_disc = torch.optim.Adam(
    discriminator.parameters(),
    lr=cfg.lr,
    betas=cfg.betas,
)

scaler_ae = GradScaler(enabled=torch.cuda.is_available())
scaler_disc = GradScaler(enabled=torch.cuda.is_available())

out_dir = Path(cfg.out_dir)
sample_dir = out_dir / "samples"
ckpt_dir = out_dir / "checkpoints"

sample_dir.mkdir(parents=True, exist_ok=True)
ckpt_dir.mkdir(parents=True, exist_ok=True)

sum(p.numel() for p in autoencoder.parameters()) / 1e6, sum(p.numel() for p in discriminator.parameters()) / 1e6
```

```python
def generator_autoencoder_loss(
    x: torch.Tensor,
    x_hat: torch.Tensor,
    vq_loss: torch.Tensor,
    global_step: int,
):
    """
    VQGAN autoencoder-side loss.

    Paper-style structure:

        L = L_rec
            + L_codebook
            + lambda * L_GAN

    where:

        L_rec includes pixel reconstruction and perceptual reconstruction.

        L_codebook =
            ||sg[z] - z_q||^2
            + beta * ||z - sg[z_q]||^2

        L_GAN is the generator adversarial loss.

        lambda is adaptive, based on gradient norm ratio between reconstruction
        loss and GAN loss with respect to the decoder's last layer.
    """

    pixel_loss = torch.mean(torch.abs(x - x_hat))

    perceptual_loss = torch.mean(perceptual_loss_fn(x, x_hat))

    nll_loss = (
        cfg.reconstruction_weight * pixel_loss
        + cfg.perceptual_weight * perceptual_loss
    )

    logits_fake = discriminator(x_hat)
    g_loss = vanilla_g_loss(logits_fake)

    disc_factor = adopt_weight(
        cfg.disc_factor,
        global_step,
        threshold=cfg.disc_start_step,
        value=0.0,
    )

    if disc_factor > 0:
        adaptive_weight = calculate_adaptive_weight(
            nll_loss=nll_loss,
            g_loss=g_loss,
            last_layer=autoencoder.get_last_layer(),
            max_weight=cfg.adaptive_weight_max,
        )
    else:
        adaptive_weight = torch.tensor(0.0, device=x.device)

    gan_loss = disc_factor * cfg.disc_weight * adaptive_weight * g_loss

    total_loss = nll_loss + cfg.codebook_weight * vq_loss + gan_loss

    logs = {
        "total_loss": total_loss.detach(),
        "pixel_loss": pixel_loss.detach(),
        "perceptual_loss": perceptual_loss.detach(),
        "nll_loss": nll_loss.detach(),
        "vq_loss": vq_loss.detach(),
        "g_loss": g_loss.detach(),
        "adaptive_weight": adaptive_weight.detach(),
        "disc_factor": torch.tensor(disc_factor, device=x.device),
        "gan_loss": gan_loss.detach(),
    }

    return total_loss, logs
```

```python
def discriminator_loss(
    x: torch.Tensor,
    x_hat: torch.Tensor,
    global_step: int,
):
    disc_factor = adopt_weight(
        cfg.disc_factor,
        global_step,
        threshold=cfg.disc_start_step,
        value=0.0,
    )

    logits_real = discriminator(x.detach())
    logits_fake = discriminator(x_hat.detach())

    d_loss = disc_factor * hinge_d_loss(logits_real, logits_fake)

    logs = {
        "disc_loss": d_loss.detach(),
        "logits_real": logits_real.mean().detach(),
        "logits_fake": logits_fake.mean().detach(),
        "disc_factor": torch.tensor(disc_factor, device=x.device),
    }

    return d_loss, logs
```

```python
@torch.no_grad()
def save_reconstruction_grid(global_step: int, max_images: int = 8):
    autoencoder.eval()

    original_0_1, x = next(iter(vqgan_loader))
    x = x[:max_images].to(device)

    x_hat, _, indices = autoencoder(x)

    grid = make_grid(
        torch.cat([to_0_1(x), to_0_1(x_hat)], dim=0),
        nrow=max_images,
        padding=2,
    )

    path = sample_dir / f"step_{global_step:08d}.png"
    TF.to_pil_image(grid.cpu()).save(path)

    print(f"Saved sample grid: {path}")
    print(f"Codebook entries used in sample: {indices.unique().numel()} / {cfg.n_embed}")

    autoencoder.train()
```

```python
def save_checkpoint(global_step: int, epoch: int):
    ckpt = {
        "global_step": global_step,
        "epoch": epoch,
        "cfg": cfg.__dict__,
        "autoencoder": autoencoder.state_dict(),
        "discriminator": discriminator.state_dict(),
        "opt_ae": opt_ae.state_dict(),
        "opt_disc": opt_disc.state_dict(),
        "scaler_ae": scaler_ae.state_dict(),
        "scaler_disc": scaler_disc.state_dict(),
    }

    path = ckpt_dir / f"vqgan_step_{global_step:08d}.pt"
    torch.save(ckpt, path)
    print(f"Saved checkpoint: {path}")


def load_checkpoint(path: str | Path):
    path = Path(path)
    ckpt = torch.load(path, map_location=device)

    autoencoder.load_state_dict(ckpt["autoencoder"])
    discriminator.load_state_dict(ckpt["discriminator"])
    opt_ae.load_state_dict(ckpt["opt_ae"])
    opt_disc.load_state_dict(ckpt["opt_disc"])
    scaler_ae.load_state_dict(ckpt["scaler_ae"])
    scaler_disc.load_state_dict(ckpt["scaler_disc"])

    print(f"Loaded checkpoint: {path}")
    return ckpt["global_step"], ckpt["epoch"]
```

```python
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
from PIL import Image
import numpy as np


def tensor_neg1_1_to_display_image(x: torch.Tensor) -> torch.Tensor:
    """
    Convert a tensor image from [-1, 1] to display-ready [0, 1].

    Input:
        x shape: (3, H, W)

    Output:
        x_display shape: (H, W, 3), clipped to [0, 1]
    """
    x = x.detach().float().cpu()
    x = (x + 1.0) / 2.0
    x = x.clamp(0.0, 1.0)
    x = x.permute(1, 2, 0)
    return x


def make_lanczos_downsample_display_image(
    x: torch.Tensor,
    latent_height: int = 64,
    latent_width: int = 48,
) -> np.ndarray:
    """
    Convert an image tensor from [-1, 1] to [0, 1],
    then Lanczos-downsample it to latent resolution.

    Input:
        x shape: (3, 256, 192)

    Output:
        image array shape: (64, 48, 3), values in [0, 1]

    Matplotlib will display it at the same subplot height as the other images.
    """
    img = tensor_neg1_1_to_display_image(x)
    img_np = (img.numpy() * 255.0).round().astype(np.uint8)

    pil_img = Image.fromarray(img_np, mode="RGB")
    pil_small = pil_img.resize(
        (latent_width, latent_height),
        resample=Image.Resampling.LANCZOS,
    )

    small_np = np.asarray(pil_small).astype(np.float32) / 255.0
    return small_np


def latent_3ch_to_display_image(z: torch.Tensor) -> torch.Tensor:
    """
    Visualize a 3-channel latent tensor as an RGB-like image.

    Input:
        z shape: (3, 64, 48)

    Output:
        z_display shape: (64, 48, 3), min-max scaled to [0, 1]
    """
    z = z.detach().float().cpu()

    z_min = z.min()
    z_max = z.max()

    if (z_max - z_min) < 1e-8:
        z = torch.zeros_like(z)
    else:
        z = (z - z_min) / (z_max - z_min)

    z = z.clamp(0.0, 1.0)
    z = z.permute(1, 2, 0)
    return z


@torch.no_grad()
def show_training_visualization(
    x: torch.Tensor,
    x_hat: torch.Tensor,
    z_q: torch.Tensor,
    global_step: int,
    epoch: int,
    batch_idx: int,
):
    """
    Show:
      1. first input image from current batch
      2. Lanczos-downsampled input at 64 x 48
      3. reconstruction
      4. 64 x 48 x 3 quantized latent visualization
    """
    input_img = tensor_neg1_1_to_display_image(x[0])
    lanczos_img = make_lanczos_downsample_display_image(
        x[0],
        latent_height=z_q.shape[-2],
        latent_width=z_q.shape[-1],
    )
    recon_img = tensor_neg1_1_to_display_image(x_hat[0])
    latent_img = latent_3ch_to_display_image(z_q[0])

    clear_output(wait=True)

    fig, axes = plt.subplots(1, 4, figsize=(16, 4))

    axes[0].imshow(input_img)
    axes[0].set_title("Input image")
    axes[0].axis("off")

    # interpolation="nearest" keeps the 64x48 pixels visually distinct
    # when matplotlib enlarges it to subplot size.
    axes[1].imshow(lanczos_img, interpolation="nearest")
    axes[1].set_title("Lanczos downsample\n64 x 48")
    axes[1].axis("off")

    axes[2].imshow(recon_img)
    axes[2].set_title("Reconstruction")
    axes[2].axis("off")

    axes[3].imshow(latent_img, interpolation="nearest")
    axes[3].set_title("Quantized latent\nmin-max scaled")
    axes[3].axis("off")

    plt.suptitle(
        f"epoch={epoch} batch={batch_idx} step={global_step} | "
        f"latent shape: {tuple(z_q[0].shape)}"
    )

    plt.tight_layout()
    plt.show()
```
